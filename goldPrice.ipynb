## model training

```python
x = df[["USD_INR"]]   # DataFrame (2D)
y = df["GOLD PRICE"]
```

```python
x
```

```python
y
```

```python
## split the train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train[["USD_INR"]])
x_test_scaled  = scaler.transform(x_test[["USD_INR"]])
```

```python
print(type(x_train))
print(x_train.head())

```

```python
## model train
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train_scaled,y_train)
```

```python
#m-slope
regression.coef_[0]
```

```python
#b-intercept
regression.intercept_
```

```python
#y=mx+b
m=regression.coef_[0]
b=regression.intercept_
```

```python
m,b
```

```python
x_train_predic=regression.predict(x_train_scaled)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_train,y_train)
plt.plot(x_train,x_train_predic,color='r')
plt.xlabel("USD_INR")
plt.ylabel("GOLD PRICE")
plt.title("2025 GOLD PRICE VS USD")
```

```python
x_test_predict=regression.predict(x_test_scaled)
```

```python
x_test_predict
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,x_test_predict)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
## hyperparameter optimization
param_space = {
    "copy_X": [True, False],
    "fit_intercept": [True, False],
    "n_jobs": [1, 5, 10, 15, None],
    "positive": [True, False]
}
```

```python
search=RandomizedSearchCV(regression,param_space,n_iter=50,cv=5)
```

```python
search.fit(x_train,y_train)
```

```python
search.best_params_
```

```python


tuned_model = LinearRegression(
    positive=True,
    n_jobs=1,
    fit_intercept=True,
    copy_X=True
)
```

```python
tuned_model.fit(x_train_scaled,y_train)
```

```python
tuned_model.coef_
```
